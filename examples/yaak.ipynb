```python
from hydra import compose, initialize
from hydra.utils import instantiate
```

#### Compose a [hydra](https://hydra.cc) config:

```python
! just generate-config
CONFIG_PATH = "../config"
DATA_DIR = "/nasa/drives/yaak/data"
DATASET = "yaak/train-tiny.yaml"
LOGGER = f"console"

with initialize(version_base=None, config_path=CONFIG_PATH):
    cfg = compose(
        "visualize",
        overrides=[
            "dataloader=unbatched",
            f"dataset={DATASET}",
            f"logger={LOGGER}",
            f"+paths.data={DATA_DIR}",
            "+paths.rbyte.cache=.cache",
        ],
    )
```

#### Instantiate the dataloader:

```python
dataloader = instantiate(cfg.dataloader)
```

```python
df = dataloader.dataset.samples

for i, frame_idx_list in enumerate(df['meta/ImageMetadata.cam_front_left/frame_idx']):
    deltas = [j - i for i, j in zip(frame_idx_list, frame_idx_list[1:])]
    if any(delta != 10 for delta in deltas):
        print(f"{i} : {list(frame_idx_list)}")
```
